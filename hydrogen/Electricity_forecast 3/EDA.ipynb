```python
plt.figure(figsize = (20, 15))
maes = []
for date in dates: 
    dam = dam_groups_dfs[6][dam_groups_dfs[6].OPR_DT == date][["MW"]]
    mean_h = groups_dfs[6][groups_dfs[6].OPR_DT == date][["MW"]].groupby(groups_dfs[6][groups_dfs[6].OPR_DT == date][["MW"]].index // 12).mean()
    dam_groups_dfs[6][dam_groups_dfs[6].OPR_DT == date][["MW"]].plot()
    groups_dfs[6][groups_dfs[6].OPR_DT == date][["MW"]].groupby(groups_dfs[6][groups_dfs[6].OPR_DT == date][["MW"]].index // 12).mean().plot()
    maes.append(mape(mean_h, dam))
    print(maes)
print(sum(maes)/len(maes))
```
